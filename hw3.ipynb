```python
from collections import defaultdict
from itertools import product
import numpy as np
```

```python
LETTERS_LIST = [chr(ord('а') + i) for i in range(ord('я') - ord('а') + 1)]
LETTERS_LIST.append(' ')
print(LETTERS_LIST)
```

```python
LETTERS_SET = set(LETTERS_LIST)
```

```python
def get_freq_dict(text, extra_gram_len=0):
  freq_dict = defaultdict(int)
  for pos in range(len(text) - extra_gram_len):
    freq_dict[text[pos:pos + extra_gram_len + 1]] += 1
  all_combs = [''.join(el) for el in product(LETTERS_LIST, repeat=extra_gram_len+1)]
  for gram in all_combs:
    freq_dict[gram] += 1
  return freq_dict

def text_fix(text_part):
  result = []
  for let in text_part:
    if let.lower() in LETTERS_SET:
      result.append(let.lower())
  return ''.join(result)

def encoder(text, tr_map, extra_gram_len=0):
  result = []
  for pos in range(len(text) - extra_gram_len):
    result.append(tr_map[text[pos:pos+ extra_gram_len + 1]])
  return ''.join(result)

def decoder(text, freq_dict, extra_gram_len):
  _get_freq_list = lambda sort_dict: [el[0] for el in sorted(list(sort_dict.items()),
                                                             key=lambda x: x[1])]
  decode_freq_str = _get_freq_list(freq_dict)
  text_freq_dict = get_freq_dict(text, extra_gram_len)
  text_freq_str = _get_freq_list(text_freq_dict)
  tr_map = {key: val for (key, val) in zip(text_freq_str, decode_freq_str)}
  return encoder(text, tr_map, extra_gram_len)
```

```python
from tqdm import tqdm
def get_log_likelihood(text: str, tr_dict_basic: dict, tr_dict_bigr: dict) -> float:
    decode_text = encoder(text, tr_dict_basic, extra_gram_len=0) 
    log_likelihood = 0
    for i in range(len(decode_text) - 2):
        log_likelihood += np.log(tr_dict_bigr[decode_text[i:i+2]])
    return log_likelihood

def get_new_dict_translation(tr_dict_basic: dict) -> dict:
  dict_cp = tr_dict_basic.copy()
  s1, s2 = np.random.choice(list(dict_cp.keys()), size=2, replace=False)
  dict_cp[s1], dict_cp[s2] = dict_cp[s2], dict_cp[s1]
  return dict_cp

def log_accept(l1, l2) -> bool:
    return True if l2 > l1 else np.random.rand() < (np.exp(l2-l1))


def mcmc_decode_text(text: str, tr_dict_basic: dict, tr_dict_bigr: dict, it_num=4000) -> str:
  _get_freq_list = lambda sort_dict: [el[0] for el in sorted(list(sort_dict.items()),
                                                              key=lambda x: x[1])]
  decode_freq_str = _get_freq_list(tr_dict_basic)

  text_freq_dict = get_freq_dict(text, extra_gram_len=0)
  text_freq_str = _get_freq_list(text_freq_dict)
  basic_tr_map = {key: val for (key, val) in zip(text_freq_str, decode_freq_str)}

  curr_l = get_log_likelihood(text, basic_tr_map, tr_dict_bigr)
  for i in tqdm(range(it_num)):
    new_tr_map = get_new_dict_translation(basic_tr_map)
    new_l = get_log_likelihood(text, new_tr_map, tr_dict_bigr)
    basic_tr_map, curr_l = (new_tr_map, new_l) if log_accept(curr_l, new_l) else (basic_tr_map, curr_l)

  return encoder(text, basic_tr_map)  # text.translate(dict_translation)




```
